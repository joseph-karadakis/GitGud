```python
#Dependencies
import pandas as pd
from pprint import pprint
import matplotlib.pyplot as plt
import numpy as np

```

```python
#Read from "SpotifyTop200_Final.csv" for all Data Vizzes
spot_top_200 = pd.read_csv("Output/SpotifyTop200_Final.csv")
spot_top_200
```

```python
#test DataFrame condensation with small sample.
jan_2021 = spot_top_200.loc[spot_top_200['year']==2021]
jan_2021 = jan_2021.loc[jan_2021['month']==1]
jan_2021
```

```python
#test scatter plot w/ test DataFrame
jan_2021.plot.scatter(x = 'danceability', y = 'energy')
```

```python
#create Antithesis function that will determine the inverse of a given dataframe's song attributes
def antithesis(df):
    print('Regular Averages:')
    print('-----------------')
    print(f"Danceability: {df['danceability'].mean()}")
    print(f"Energy: {df['energy'].mean()}")
    print(f"Instrumentalness: {df['instrumentalness'].mean()}")
    print(f"Speechiness: {df['speechiness'].mean()}")
    print(f"Acousticness: {df['acousticness'].mean()}")
    print(f"Liveness: {df['liveness'].mean()}")
    print(f"Valence: {df['valence'].mean()}\n")

    print('Antithesis Averages:')
    print('-----------------')
    print(f"Danceability: {1-df['danceability'].mean()}")
    print(f"Energy: {1-df['energy'].mean()}")
    print(f"Instrumentalness: {1-df['instrumentalness'].mean()}")
    print(f"Speechiness: {1-df['speechiness'].mean()}")
    print(f"Acousticness: {1-df['acousticness'].mean()}")
    print(f"Liveness: {1-df['liveness'].mean()}")
    print(f"Valence: {1-df['valence'].mean()}")


```

```python
#testing antithesis function
antithesis(jan_2021)
```

```python
#condensing dataframe to numerical, non-boolean columns that will create Data Vizzes
df_to_plot = spot_top_200[['year','month','speechiness','instrumentalness','acousticness','liveness','valence','danceability']]
df_to_plot
```

```python
#testing w/ only danceability columns for all months and years
year_means = (1-spot_top_200.groupby(['year','month'])['danceability'].mean())

year_means
```

```python
#graphing test danceability year_means df
year_means.plot(kind = 'line', figsize=[10,10], grid = True ,rot= 90 ,xlabel= 'Date (by YYYY,D)', ylabel= 'Danceability')

```

```python
#creates "Antithesis" averages df from condensed df_to_plot. Based off antithesis function from earlier.
plot_means = (1-df_to_plot.groupby(['year','month']).mean())
plot_means
```

```python
#creates a list of dates to use as xlabels
dates = []
for d in range(2017,2022):
    for i in range(1,13):
        dates.append(f"{i}/{d}")

plot_means.reset_index()
plot_means['xlabel']= dates
plot_means = plot_means.set_index('xlabel')
plot_means.head()
```

```python
#creates final data viz w/ all float "Antithesis" column averages, and saves it as a PNG.
graph = plot_means.plot(kind='line',figsize=(20,6),xlabel= 'Date',title='Inverse of Top 200 Song Attributes',grid=True)

graph.grid(alpha = .25)
graph.legend(loc = 'lower left')
graph.set_ylim(ymin=0)


plt.savefig("InverseTop200SongAtt.png")
```
