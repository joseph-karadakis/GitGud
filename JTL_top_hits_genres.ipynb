```python
#dependencies
import os
import csv
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# File to Load 
top_hits = Path("resources/top_hits_2000-2019.csv")
```

```python
#Data File and store into Pandas DataFrames
top_hits_df = pd.read_csv(top_hits)
top_hits_df
```

```python
#sort the year column in ascending order
sorted_df = top_hits_df.sort_values('year', ascending=True)
sorted_df
```

```python
#get unique genres 
unique_genres = sorted_df['genre'].unique()
unique_genres
```

```python

#count the number of genres
num_genres = len(unique_genres)
num_genres
```

```python
#split genre columns into separate columns
top_hits_df[["genre 1", "genre 2", "genre 3", "genre 4"]] = top_hits_df.genre.str.split(",", expand=True)
top_hits_df
```

```python
#check unique genre in genre 1
unique_genre_1 = top_hits_df['genre 1'].unique()
unique_genre_1

```

```python
#rename 'set()' to "Unknown"
top_hits_clean = top_hits_df.replace('set()', 'unknown')
top_hits_clean
```

```python

#check set() has been properly renamed
unique_genre = top_hits_clean['genre 1'].unique()
unique_genre
```

```python
#create new dataframe with only columns year, genre, and songs
new_df = pd.DataFrame(top_hits_clean[['year', 'genre 1','song']])
new_df

#filter dataframe for years 2000 to 2020
year_filter = new_df[(new_df['year'] >= 2000) & (new_df['year'] <= 2020)]
year_filter
```

```python
# Group the data by year and genre, and calculate the count for each genre
top_year_genre = year_filter.groupby(['year', 'genre 1']).count()

# Pivot the data to get the counts of songs by year and genre
pivoted_genres = top_year_genre.pivot_table(index='year', columns='genre 1', values='song', aggfunc='sum')
pivoted_genres

pivoted_genres.plot(kind='bar', stacked=True, figsize=(10,6))
plt.xlabel('Year')
plt.ylabel('Count of Genres')
plt.title('Popularity of Genres from 2000 to 2020')
plt.legend(loc = 'right')
```
