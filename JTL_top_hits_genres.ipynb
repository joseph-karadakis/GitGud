```python
#dependencies
import os
import csv
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# File to Load 
top_hits = Path("resources/top_hits_2000-2019.csv")
```

```python
#Data File and store into Pandas DataFrames
top_hits_df = pd.read_csv(top_hits)
top_hits_df
```

```python
#sort the year column in ascending order
sorted_df = top_hits_df.sort_values('year', ascending=True)
sorted_df
```

```python
#get unique genres 
unique_genres = sorted_df['genre'].unique()
unique_genres
```

```python

#count the number of genres
num_genres = len(unique_genres)
num_genres
```

```python
#split genre columns into separate columns
top_hits_df[["genre 1", "genre 2", "genre 3", "genre 4"]] = top_hits_df.genre.str.split(",", expand=True)
top_hits_df
```

```python
#check unique genre in genre 1
unique_genre_1 = top_hits_df['genre 1'].unique()
unique_genre_1

```

```python
#rename 'set()' to "Unknown"
top_hits_clean = top_hits_df.replace('set()', 'unknown')
top_hits_clean
```

```python

#check set() has been properly renamed
unique_genre = top_hits_clean['genre 1'].unique()
unique_genre
```

```python
# Group the data by year and genre, and calculate the count for each genre
top_year_genre = top_hits_clean.groupby(['year', 'genre 1']).count()
top_year_genre
```

```python

# Plot the genre popularity over time using a stacked bar chart
stacked_top_genres = top_year_genre.plot(kind='bar', stacked=True)
stacked_top_genres.set_xlabel('Year')
stacked_top_genres.set_ylabel('Genre Count')
stacked_top_genres.set_title('Popularity of Top Genres over Time')
plt.show()
```
