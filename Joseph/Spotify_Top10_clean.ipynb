```python
#Dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

# access data
path= 'Output/Spotify_USA_Top10_raw.csv'

#create dataframe
top10 = pd.read_csv(path)
top10.head()
```

```python
#get list of keys
# Get a list of all unique values in the 'key' column
key_list = top10['key'].unique()

# Display the list of unique values
print(key_list)
```

```python
#Convert key to Alpha
key_letters = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']
top10['key'] = top10['key'].astype(int).apply(lambda x: key_letters[x % 12] if x != -1 else 'Unknown Key')

top10.head(100)
```

```python
#get list of keys in Alpha
# Get a list of all unique values in the 'key' column
key_list_alpha = top10['key'].unique()

# Display the list of unique values
print(key_list_alpha)
```

```python
#Convert mode to Alpha
mode_alpha = ['Maj', 'Min']
top10['mode'] = top10['mode'].astype(int).apply(lambda x: mode_alpha[x % 12] if x != -1 else 'Unknown Mode')

top10.head(100)
```

```python
#get list of modes in Alpha
# Get a list of all unique values in the 'key' column
mode_list_alpha = top10['mode'].unique()

# Display the list of unique values
print(mode_list_alpha)
```

```python
#Round Up Tempo
import math

# Round up the 'tempo' column to the nearest whole number
top10['tempo'] = top10['tempo'].apply(lambda x: math.ceil(x))
top10.head()
```

```python
#Convert Duration
# convert milliseconds to minutes
top10['duration_mins'] = top10['duration_ms'] / 60000
top10['duration_mins']=top10['duration_mins'].round(2)

# drop old column
top5=top10.drop('duration_ms',axis=1)
top5.head()
```

```python
#Define Valence in Alpha
def valence_category(valence):
    if valence >= 0.8:
        return "Extremely Positive"
    elif valence >= 0.6:
        return "Positive"
    elif valence >= 0.4:
        return "Neutral"
    elif valence >= 0.2:
        return "Negative"
    else:
        return "Extremely Negative"

top10['valence_category'] = top10['valence'].apply(valence_category)
top10.head(100)
```

```python
#Define Danceability in Alpha
def danceability_category(danceability):
    if danceability >= 0.8:
        return "Most Danceable"
    elif danceability >= 0.6:
        return "More Danceable"
    elif danceability >= 0.4:
        return "Danceable"
    elif danceability >= 0.2:
        return "Less Danceable"
    else:
        return "Least Danceable"

top10['danceabity_category'] = top10['danceability'].apply(danceability_category)
top10.head(100)
```

```python
#Define Energy in Alpha
def energy_category(energy):
    if energy >= 0.8:
        return "Extremely Intense"
    elif energy >= 0.6:
        return "Intense"
    elif energy >= 0.4:
        return "Neutral"
    elif energy >= 0.2:
        return "Calm"
    else:
        return "Extremely Calm"

top10['energy_category'] = top10['energy'].apply(energy_category)
top10.head(100)
```

```python
# Extract the year and month into separate column
top10['date'] = pd.to_datetime(top10['date'])
top10['month_year'] = top10['date'].dt.to_period('M')  # Format: YYYY-MM

# Drop the original date column if desired
top10.drop(columns=['date'], inplace=True)

top10.head()
```

```python
print(top10.columns)
```

```python
#Sort Columns
# Define the desired column order
column_order = ['title', 'artist', 'month_year', 'tempo', 'duration_mins', 'valence_category',
                'danceabity_category', 'energy_category', 'key', 'mode', 'rank', 'region', 'chart',
                'trend', 'streams', 'danceability', 'energy', 'loudness', 'speechiness', 
                'acousticness', 'instrumentalness', 'liveness', 'valence', 'url', 'Spotify Track Id']

# Reindex the DataFrame using the desired column order
top10 = top10.reindex(columns=column_order)

# Display the resulting DataFrame
top10.head(100)
```

```python
# Create a directory named "Output" if it doesn't exist
Path("Output").mkdir(exist_ok=True)

# Save the DataFrame as a CSV in the "Output" directory
top10.to_csv("Output/Spotify_USA_Top10_clean.csv", index=False)
```
