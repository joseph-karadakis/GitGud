```python
#dependencies
import os
import csv
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# File to Load 
top_hits = Path("Resources/top_hits_2000-2019.csv")
```

```python
#Data File and store into Pandas DataFrames
top_hits_df = pd.read_csv(top_hits)
top_hits_df
```

```python
#sort the year column in ascending order
sorted_df = top_hits_df.sort_values('year', ascending=True)
```

```python
#check get unique genres 
unique_genres = sorted_df['genre'].unique()
unique_genres
```

```python

#count the number of genres
num_genres = len(unique_genres)
num_genres
```

```python
#split genre columns into separate columns
top_hits_df[["genre 1", "genre 2", "genre 3", "genre 4"]] = top_hits_df.genre.str.split(",", expand=True)
top_hits_df
```

```python
#check unique genre in genre 1
unique_genre_1 = top_hits_df['genre 1'].unique()
unique_genre_1

```

```python
#rename 'set()' to "Unknown"
top_hits_clean = top_hits_df.replace('set()', 'unknown')
top_hits_clean
```

```python

#check set() has been properly renamed
unique_genre = top_hits_clean['genre 1'].unique()
unique_genre
```

```python
#create new dataframe with only columns year, genre, and songs
new_df = pd.DataFrame(top_hits_clean[['year', 'genre 1','song']])
new_df

#filter dataframe for years 2000 to 2020
year_filter = new_df[(new_df['year'] >= 2000) & (new_df['year'] <= 2019)]
year_filter
```

```python
# Group the data by year and genre and calculate the count for each genre
top_year_genre = year_filter.groupby(['year', 'genre 1']).count()
top_year_genre = top_year_genre.reset_index()
top_year_genre
```

```python
# Group the data by year and genre, and calculate the count for each genre
top_year_genre = year_filter.groupby(['year', 'genre 1']).count()
# Pivot the data to get the counts of songs by year and genre
pivoted_genres = top_year_genre.pivot_table(index='year', columns='genre 1', values='song', aggfunc='sum')

#plot the data
pivoted_genres.plot(kind='bar', stacked=True, figsize=(15,10))
plt.xlabel('Year')
plt.ylabel('Count of Genres')
plt.title('Popularity of Genres from 2000 to 2019')
plt.legend(loc = 'best')
```

```python
#filter to contain only 2008
year_2008 = new_df[(new_df['year'] == 2008)]

#group by genre then get total song count in each genre
year_2008_groupby = year_2008.groupby('genre 1').count()

#drop year column
year_2008_clean = year_2008_groupby.drop(columns=['year'])
year_2008_clean
```

```python
#check data type for year_2008_clean
year_2008_clean.dtypes
```

```python
#pie chart of generes in 2008
labels = ['Dance/Electronic', 'R&B','Hip hop','Pop', 'Rock']
colors = ["orange", "blue", "pink", 'green', "red"]
sizes =["2","2","35","47","11"]
plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%", startangle=140)
plt.title('Popular Genres in 2008')
```
