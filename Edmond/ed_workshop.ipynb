```python
#on gitbash or command prompt: pip install pytrends
#importing dependencies
from pytrends.request import TrendReq
#connects to google and assigns it to variable pt
pt = TrendReq(hl = 'en-US', tz = 480)
```

```python
pt.realtime_trending_searches(pn = 'US')
```

```python
pt.build_payload(kw_list=['insurrection'],geo='US',timeframe = ['2020-07-06 2021-07-06'])
result = pt.interest_over_time()
result.plot(kind= 'line')
```
