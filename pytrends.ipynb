```python
from pytrends.request import TrendReq
pt = TrendReq(hl = 'en-US', tz = 360)
```

```python
pt.build_payload(kw_list=['Ron DeSantis'], geo = 'US', timeframe= 'today 1-m')
interest = pt.interest_over_time()
interest
```

```python
pt.suggestions('desantis')
```
